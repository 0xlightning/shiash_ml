```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Heart_disease.csv')
df
```

```python
df.isnull().sum()
```

```python
df.skew()
```

```python
df.glucose = df.glucose.fillna(df.glucose.median())
df.education = df.education.fillna(df.education.median())
df.cigsPerDay = df.cigsPerDay.fillna(df.cigsPerDay.median())
df.BPMeds = df.BPMeds.fillna(df.BPMeds.median())
df.BMI = df.BMI.fillna(df.BMI.median())
df.totChol = df.totChol.fillna(df.totChol.median())
df.heartRate = df.heartRate.fillna(df.heartRate.median())

df.isnull().sum()
```

```python
def gender_mapping(val):
    if val==0:
        val='Female'
    else:
        val='Male'
    return val
```

```python
def fun1(val):
    if val==1:
        val='Yes'
    else:
        val='No'
    return val
```

```python
def education1(edu):
    if edu==1:
        edu='High School'
    elif edu==2:
        edu='Some College'
    elif edu==3:
        edu='Bachelor'
    elif edu==4:
        edu='Master'
    return edu
```

```python
df["BPMeds"].unique()
```

```python
df["male"]=df["male"].apply(gender_mapping)
df['education']=df['education'].apply(education1)
df['currentSmoker']=df['currentSmoker'].apply(fun1)
df["BPMeds"]=df["BPMeds"].apply(fun1)
df["prevalentStroke"]=df["prevalentStroke"].apply(fun1)
df["diabetes"]=df["diabetes"].apply(fun1)
df["prevalentHyp"]=df["prevalentHyp"].apply(fun1)
df 
```

```python
df["male"]=df["male"].astype("object")
df["education"]=df["education"].astype("object")
df["currentSmoker"]=df["currentSmoker"].astype("object")
df["BPMeds"]=df["BPMeds"].astype("object")
df["prevalentStroke"]=df["prevalentStroke"].astype("object")
df["diabetes"]=df["diabetes"].astype("object")
df["prevalentHyp"]=df["prevalentHyp"].astype("object")
# df["CVD"]=df["CVD"].astype("object")

df.info()
```

```python
# Example of a box plot for 'age' vs 'CVD'
plt.figure(figsize=(8, 6))
sns.boxplot(x='CVD', y='age', data=df)
plt.title('Age vs Cardiovascular Disease (CVD)')
plt.xlabel('CVD')
plt.ylabel('Age')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df1_num_sc = pd.DataFrame(sc.fit_transform(num_df), columns=num_df.columns)
df1_num_sc
```

```python
df1_pre = pd.concat([df1_num_sc, dummy], axis=1)
df1_pre
```

```python
# Split the dataset into features (X) and target (Y)
X = df1_pre.drop('CVD', axis=1)
Y = df['CVD']
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
y_train.sort_index(ascending=True,inplace=True)
x_train.sort_index(ascending=True,inplace=True)
x_test.sort_index(ascending=True,inplace=True)
y_test.sort_index(ascending=True,inplace=True)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100,random_state=0)
rfc
```

```python
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
print("The accuracy score of the model is",acc)
```
