```python
import pandas as pd
import numpy as np
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Heart_disease.csv')
df
```

```python
df.isnull().sum()
```

```python
df.skew()
```

```python
df.glucose = df.glucose.fillna(df.glucose.median())
df.education = df.education.fillna(df.education.median())
df.cigsPerDay = df.cigsPerDay.fillna(df.cigsPerDay.median())
df.BPMeds = df.BPMeds.fillna(df.BPMeds.median())
df.BMI = df.BMI.fillna(df.BMI.median())
df.totChol = df.totChol.fillna(df.totChol.median())
df.heartRate = df.heartRate.fillna(df.heartRate.median())

df.isnull().sum()
```

```python
def gender_mapping(val):
    if val==0:
        val='Female'
    else:
        val='Male'
    return val
```

```python
def fun1(val):
    if val==1:
        val='Yes'
    else:
        val='No'
    return val
```

```python
def education1(edu):
    if edu==1:
        edu='High School'
    elif edu==2:
        edu='Some College'
    elif edu==3:
        edu='Bachelor'
    elif edu==4:
        edu='Master'
    return edu
```

```python
df["BPMeds"].unique()
```

```python
df["male"]=df["male"].apply(gender_mapping)
df['education']=df['education'].apply(education1)
df['currentSmoker']=df['currentSmoker'].apply(fun1)
df["BPMeds"]=df["BPMeds"].apply(fun1)
df["prevalentStroke"]=df["prevalentStroke"].apply(fun1)
df["diabetes"]=df["diabetes"].apply(fun1)
df["prevalentHyp"]=df["prevalentHyp"].apply(fun1)
df 
```

```python
df["male"]=df["male"].astype("object")
df["education"]=df["education"].astype("object")
df["currentSmoker"]=df["currentSmoker"].astype("object")
df["BPMeds"]=df["BPMeds"].astype("object")
df["prevalentStroke"]=df["prevalentStroke"].astype("object")
df["diabetes"]=df["diabetes"].astype("object")
df["prevalentHyp"]=df["prevalentHyp"].astype("object")
# df["CVD"]=df["CVD"].astype("object")

df.info()
```

```python
# Split the dataset into features (X) and target (Y)
X = df1_pre.drop('CVD', axis=1)
Y = df['CVD']
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
y_train.sort_index(ascending=True,inplace=True)
x_train.sort_index(ascending=True,inplace=True)
x_test.sort_index(ascending=True,inplace=True)
y_test.sort_index(ascending=True,inplace=True)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
logreg=LogisticRegression()
logreg.fit(x_train, y_train)
y_pred=logreg.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
from sklearn.metrics import precision_score, recall_score
precision = precision_score(y_test, y_pred)
precision
```

```python
recall=recall_score(y_test,y_pred)
recall
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
conf_matrix = pd.DataFrame(data = cm,
columns = ['Predicted:0(Safe)', 'Predicted:1(Death)'],
index =['Actual:0(Safe)', 'Actual:1(Death)'])

plt.figure(figsize = (8, 5))
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Greens")

plt.show()

print('The details for confusion matrix is =')
print (classification_report(y_test, y_pred))
```
