```python
import pandas as pd
import numpy as np
```

```python
# # A reimbursement is a compensation paid by an organisation or a company for any out-of-pocket inverstments 
# made by an employee  or an overpayment made by a customer or
# any other party that the company deals with

# Some common examples of reimbursemnets are reimbursements of business expenses like travel 
# or food expenses, reimbursments made by insurance companies
# to the insured person for their medical bills,
# or reimbursements made to a person who makes a purshase on behalf of a third party.
```

```python
df = pd.read_csv('Emp_reimb.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df1 = df.copy()
```

```python
df1.drop(columns=['Emp_ID'],inplace=True)
```

```python
df1_num1 = df1.select_dtypes(include=['float','int'])
df1_num1
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df1_num_sc = pd.DataFrame(sc.fit_transform(df1_num1), columns =  df1_num1.columns)
df1_num_sc.head(3)
```

```python
dummy = pd.get_dummies(df1.Type_of_yr, drop_first=False)
dummy.head(7)
# take the numerical data as it is and take the catagrical data into numerical data
```

```python
df1_pre = pd.concat([df1_num_sc,dummy],axis=1)
df1_pre.head()
```

```python
x = df1_pre.drop('Total_Reimbursement', axis=1)
```

```python
y = df1.iloc[:,[-1]]
```

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train.sort_index(ascending=True, inplace=True)
x_test.sort_index(ascending=True, inplace=True)
y_train.sort_index(ascending=True, inplace=True)
y_test.sort_index(ascending=True, inplace=True)
```

```python
import statsmodels.api as sm
```

```python
MLR_model1 = sm.OLS(y_train,x_train).fit()
MLR_model1.summary()
```

```python
y_test_pred = MLR_model1.predict(x_test)
y_test_pred.count()
```

```python
y_test_pred
```

```python
y_test_pred.sort_index(ascending=True, inplace=True)
```

```python
# this was for checking the original and predicted are correct or wrong
y_test.Total_Reimbursement[33]
```

```python
# this is the predicted value
y_test_pred[33]
```

```python
y_test_pred[0:10]
```

```python
error = y_test.Total_Reimbursement-y_test_pred
error
```

```python
(y_test.Total_Reimbursement-y_test_pred)**2
```

```python
# mse value
np.sum((y_test.Total_Reimbursement-y_test_pred)**2)/len(y_test.Total_Reimbursement)
```

```python
# ML model has made accurate predictions.
# This is the mse value of test data. it is approximately 0.
# it means that ml model is a good model.
```

```python
from sklearn.metrics import r2_score
```

```python
# R2 Score is equal to 1 the prediction is very good
r2_score(y_true= y_test.Total_Reimbursement, y_pred = y_test_pred)
```
